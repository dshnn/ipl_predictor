```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
from xgboost import XGBClassifier
```

```python
d1 = pd.read_csv(r"deliveries.csv")
d2 = pd.read_csv(r"matches.csv")
d3=pd.DataFrame
```

```python
d2 = d2.sort_values(["date"]).reset_index(drop=True)
```

```python
d2.head()
common=(set(d2.team1) & set(d1.batting_team) )
for c in common:
    print(c)
```

```python
team_mapping= {
    "Delhi Daredevils" : "Delhi Capitals",
    "Royal Challengers Bangalore": " Royal Challengers Bengaluru",
    "Pune Warriors": "Rising Pune Supergiants",
    "Rising Pune Supergiant" : "Rising Pune Supergiants",
    "Gujarat Lions" : "Gujarat Titans",
    "Kings XI Punjab": "Punjab Kings"
}
```

```python
d2.team1 = d2.team1.replace(team_mapping)
```

```python
d2_replace = ["team1","team2","toss_winner","winner"]
d1_replace=["batting_team","bowling_team"]
```

```python
for i in d2_replace:
    d2[i] = d2[i].replace(team_mapping)

```

```python
for i in d1_replace:
    d1[i]= d1[i].replace(team_mapping)
```

```python
d1["cum_wickets"]=d1.groupby(["match_id","inning"])["is_wicket"].cumsum()
d1["cum_runs"]=d1.groupby(["match_id","inning"])["total_runs"].cumsum()
d1["balls"]=d1.groupby(["match_id","inning"])["ball"].cumcount()+1

d3=d1.groupby(["inning","match_id","over"],sort = False).last().reset_index()
```

```python
d2.columns
```

```python
merge = pd.merge(d2, d3, how="left", left_on="id", right_on="match_id")
```

```python
len(merge)
```

```python
h2h = {}
h2h_feature = []

for _, row in d2.iterrows():
    team1 = row["team1"]
    team2 = row["team2"]

    key = tuple(sorted([team1, team2]))

    if key not in h2h:
        h2h[key] = {team1: 0, team2: 0}

    team1_wins = h2h[key].get(team1, 0)
    team2_wins = h2h[key].get(team2, 0)

    h2h_feature.append(team1_wins - team2_wins)

    winner = row["winner"]
    if pd.notna(winner):
        h2h[key][winner] = h2h[key].get(winner, 0)+1
```

```python
d2["h2h"]=h2h_feature
```

```python
team_stats = {}
team1_win_rate = []
team2_win_rate = []

for _, row in d2.iterrows():
    team = row["team1"]
    team2 = row["team2"]
    winner = row["winner"]


    key1 = team
    key2 = team2

    if key1 not in team_stats:
        team_stats[key1] = {"won": 0, "played": 0}

    if key2 not in team_stats:
        team_stats[key2] = {"won": 0, "played": 0}

    team1_won = team_stats[key1].get("won", 0)
    team1_played = team_stats[key1].get("played", 0)

    team2_won = team_stats[key2].get("won", 0)
    team2_played = team_stats[key2].get("played", 0)

    
    if team1_played != 0:
        team1_win_rate.append(team1_won/team1_played)
    else:
        team1_win_rate.append(0)
        
    if team2_played != 0:
        team2_win_rate.append(team2_won/team2_played)
    else:
        team2_win_rate.append(0)

    team_stats[team]["played"] = team_stats[team].get("played", 0) + 1
    team_stats[team2]["played"] = team_stats[team2].get("played", 0) + 1
    
    if pd.notna(winner) and winner == team:
        team_stats[winner]["won"] = team_stats[winner].get("won", 0) + 1
    elif pd.notna(winner) and winner == team2:
        team_stats[winner]["won"] = team_stats[winner].get("won", 0) + 1
        
        
        
```

```python
d2["team1_win_rate"] = team1_win_rate
d2["team2_win_rate"] = team2_win_rate
```

```python
venue_stats = {}
team1_wr_at_venue = []
team2_wr_at_venue = []

for _, row in d2.iterrows():
    team = row["team1"]
    team2 = row["team2"]
    winner = row["winner"]
    stadium = row["venue"]


    key1 = tuple([team,stadium])
    key2 = tuple([team2,stadium])

    if key1 not in venue_stats:
        venue_stats[key1] = {"won": 0, "played": 0}

    if key2 not in venue_stats:
        venue_stats[key2] = {"won": 0, "played": 0}

    team1_won = venue_stats[key1].get("won", 0)
    team1_played = venue_stats[key1].get("played", 0)

    team2_won = venue_stats[key2].get("won", 0)
    team2_played = venue_stats[key2].get("played", 0)

    
    if team1_played != 0:
        team1_wr_at_venue.append(team1_won/team1_played)
    else:
        team1_wr_at_venue.append(0)
        
    if team2_played != 0:
        team2_wr_at_venue.append(team2_won/team2_played)
    else:
        team2_wr_at_venue.append(0)

    venue_stats[key1]["played"] = venue_stats[key1].get("played", 0) + 1
    venue_stats[key2]["played"] = venue_stats[key2].get("played", 0) + 1
    
    if pd.notna(winner) and winner == team:
        venue_stats[key1]["won"] = venue_stats[key1].get("won", 0) + 1
    elif pd.notna(winner) and winner == team2:
        venue_stats[key2]["won"] = venue_stats[key2].get("won", 0) + 1
        
        
        
```

```python
d2["team1_wr_at_venue"] = team1_wr_at_venue
d2["team2_wr_at_venue"] = team2_wr_at_venue
```

```python
venue_stats={}
avg_score=[]

for _, row in d2.iterrows():
    venue = row["venue"]

    if venue not in venue_stats:
        venue_stats[venue] = {"score": 0, "matches": 0}

    if venue_stats[venue]["matches"] != 0:
        avg_score.append(venue_stats[venue]["score"] / venue_stats[venue]["matches"])
    else:
        avg_score.append(0)

    venue_stats[venue]["score"] += row["target_runs"]
    venue_stats[venue]["matches"]+=1
```

```python
d2["avg_score_of_stadium"] = avg_score
```

```python
merge = pd.merge(d2, d3, how="left", left_on="id", right_on="match_id")
```

```python
merge['balls_remaining']=120-merge["balls"]

merge["req_runs"]=merge["target_runs"]-merge["cum_runs"]

merge["required_run_rate"] = np.where(merge["balls_remaining"]>0,
                                      merge["req_runs"] / 
                                       (merge["balls_remaining"]/6),
                                      0)

merge.loc[merge["inning"] == 1, "required_run_rate"] = 0

merge.loc[merge["inning"] == 1, "req_runs"] = 0

merge.loc[merge["inning"] == 1, "target_runs"] = 0

merge["current_run_rate"] = np.where( merge["balls"] > 0,
                                    merge["cum_runs"]/(merge["balls"]/6), 0)
merge["wicket_in_hands"] = 10 - merge["cum_wickets"]
merge["t1_won_toss"] = (merge["team1"] == merge["toss_winner"]).astype(int)
merge["toss_decision_bat"] = (merge["toss_decision"]=="bat").astype(int) 
```
