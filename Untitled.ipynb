```python
h2h = {}
h2h_feature = []

for _, row in d2.iterrows():
    team1 = row["team1"]
    team2 = row["team2"]

    key = tuple(sorted([team1, team2]))

    if key not in h2h:
        h2h[key] = {team1: 0, team2: 0}

    team1_wins = h2h[key].get(team1, 0)
    team2_wins = h2h[key].get(team2, 0)

    h2h_feature.append(team1_wins - team2_wins)

    winner = row["winner"]
    if pd.notna(winner):
        h2h[key][winner] = h2h[key].get(winner, 0)+1
```

```python
d2["h2h"]=h2h_feature
```

```python
team_stats = {}
team1_win_rate = []
team2_win_rate = []

for _, row in d2.iterrows():
    team = row["team1"]
    team2 = row["team2"]
    winner = row["winner"]


    key1 = team
    key2 = team2

    if key1 not in team_stats:
        team_stats[key1] = {"won": 0, "played": 0}

    if key2 not in team_stats:
        team_stats[key2] = {"won": 0, "played": 0}

    team1_won = team_stats[key1].get("won", 0)
    team1_played = team_stats[key1].get("played", 0)

    team2_won = team_stats[key2].get("won", 0)
    team2_played = team_stats[key2].get("played", 0)

    
    if team1_played != 0:
        team1_win_rate.append(team1_won/team1_played)
    else:
        team1_win_rate.append(0)
        
    if team2_played != 0:
        team2_win_rate.append(team2_won/team2_played)
    else:
        team2_win_rate.append(0)

    team_stats[team]["played"] = team_stats[team].get("played", 0) + 1
    team_stats[team2]["played"] = team_stats[team2].get("played", 0) + 1
    
    if pd.notna(winner) and winner == team:
        team_stats[winner]["won"] = team_stats[winner].get("won", 0) + 1
    elif pd.notna(winner) and winner == team2:
        team_stats[winner]["won"] = team_stats[winner].get("won", 0) + 1
        
        
        
```

```python
d2["team1_win_rate"] = team1_win_rate
d2["team2_win_rate"] = team2_win_rate
```

```python
venue_stats = {}
team1_wr_at_venue = []
team2_wr_at_venue = []

for _, row in d2.iterrows():
    team = row["team1"]
    team2 = row["team2"]
    winner = row["winner"]
    stadium = row["venue"]


    key1 = tuple([team,stadium])
    key2 = tuple([team2,stadium])

    if key1 not in venue_stats:
        venue_stats[key1] = {"won": 0, "played": 0}

    if key2 not in venue_stats:
        venue_stats[key2] = {"won": 0, "played": 0}

    team1_won = venue_stats[key1].get("won", 0)
    team1_played = venue_stats[key1].get("played", 0)

    team2_won = venue_stats[key2].get("won", 0)
    team2_played = venue_stats[key2].get("played", 0)

    
    if team1_played != 0:
        team1_wr_at_venue.append(team1_won/team1_played)
    else:
        team1_wr_at_venue.append(0)
        
    if team2_played != 0:
        team2_wr_at_venue.append(team2_won/team2_played)
    else:
        team2_wr_at_venue.append(0)

    venue_stats[key1]["played"] = venue_stats[key1].get("played", 0) + 1
    venue_stats[key2]["played"] = venue_stats[key2].get("played", 0) + 1
    
    if pd.notna(winner) and winner == team:
        venue_stats[key1]["won"] = venue_stats[key1].get("won", 0) + 1
    elif pd.notna(winner) and winner == team2:
        venue_stats[key2]["won"] = venue_stats[key2].get("won", 0) + 1
        
        
        
```

```python
d2["team1_wr_at_venue"] = team1_wr_at_venue
d2["team2_wr_at_venue"] = team2_wr_at_venue
```

```python
venue_stats={}
avg_score=[]

for _, row in d2.iterrows():
    venue = row["venue"]

    if venue not in venue_stats:
        venue_stats[venue] = {"score": 0, "matches": 0}

    if venue_stats[venue]["matches"] != 0:
        avg_score.append(venue_stats[venue]["score"] / venue_stats[venue]["matches"])
    else:
        avg_score.append(0)

    venue_stats[venue]["score"] += row["target_runs"]
    venue_stats[venue]["matches"]+=1
```

```python
import pandas as pd
import numpy as np

def build_feature_vector_debug(
    team1, team2, venue, inning, over,
    cum_runs, cum_wickets, target_runs, target_overs,
    t1_won_toss, model, lookup
):
    balls = over * 6
    balls_remaining = max(0, int(target_overs) * 6 - balls)
    req_runs = max(0, target_runs - cum_runs)
    required_run_rate = (req_runs / (balls_remaining / 6)
                         if balls_remaining > 0 else 99.99)
    current_run_rate = (cum_runs / (balls / 6) if balls > 0 else 0)
    wicket_in_hands = 10 - cum_wickets

    if inning == 1:
        target_runs_feat = 0
        req_runs_feat = 0
        required_run_rate_feat = 0
    else:
        target_runs_feat = target_runs
        req_runs_feat = req_runs
        required_run_rate_feat = round(required_run_rate, 4)

    h2h = lookup["h2h"].get((team1, team2), 0)
    team1_win_rate = lookup["team_wr"].get(team1, 0.5)
    team2_win_rate = lookup["team_wr"].get(team2, 0.5)
    team1_wr_at_venue = lookup["team_wr_venue"].get((team1, venue), 0.5)
    team2_wr_at_venue = lookup["team_wr_venue"].get((team2, venue), 0.5)
    avg_score_of_stadium = lookup["avg_stadium"].get(venue, 165)

    row = {
        "target_runs": target_runs_feat,
        "target_overs": target_overs,
        "h2h": h2h,
        "team1_win_rate": team1_win_rate,
        "team2_win_rate": team2_win_rate,
        "team1_wr_at_venue": team1_wr_at_venue,
        "team2_wr_at_venue": team2_wr_at_venue,
        "avg_score_of_stadium": avg_score_of_stadium,
        "inning": inning,
        "over": over,
        "ball": 6,
        "batsman_runs": 0,
        "extra_runs": 0,
        "total_runs": 0,
        "is_wicket": 0,
        "cum_wickets": cum_wickets,
        "cum_runs": cum_runs,
        "balls": balls,
        "balls_remaining": balls_remaining,
        "req_runs": req_runs_feat,
        "required_run_rate": required_run_rate_feat,
        "current_run_rate": round(current_run_rate, 4),
        "wicket_in_hands": wicket_in_hands,
        "t1_won_toss": t1_won_toss,
    }

    feature_cols = list(model.feature_names_in_)
    TEAM1_OHE_COLS = [c.replace("team1_", "") for c in feature_cols if c.startswith("team1_")]
    TEAM2_OHE_COLS = [c.replace("team2_", "") for c in feature_cols if c.startswith("team2_")]
    VENUE_OHE_COLS = [c.replace("venue_", "") for c in feature_cols if c.startswith("venue_")]

    for col in TEAM1_OHE_COLS:
        row[f"team1_{col}"] = 1 if col.strip() == team1.strip() else 0

    for col in TEAM2_OHE_COLS:
        row[f"team2_{col}"] = 1 if col.strip() == team2.strip() else 0

    for col in VENUE_OHE_COLS:
        row[f"venue_{col}"] = 1 if col.strip() == venue.strip() else 0

    return pd.DataFrame([row])


# ── run the debug test ───────────────────────────────────────────────────────
X = build_feature_vector_debug(
    team1="Mumbai Indians",
    team2="Royal Challengers Bengaluru",
    venue="Wankhede Stadium",
    inning=2,
    over=5,
    cum_runs=85,
    cum_wickets=2,
    target_runs=180,
    target_overs=20,
    t1_won_toss=1,
    model=model,
    lookup=lookup
)

X = X.reindex(columns=model.feature_names_in_, fill_value=0)

# check key features
key_cols = ['inning', 'cum_runs', 'cum_wickets', 'req_runs',
            'required_run_rate', 'current_run_rate',
            'balls_remaining', 'target_runs']
print("Key features:")
print(X[key_cols].T)

# check team encoding
team_cols = [c for c in X.columns if c.startswith('team1_') or c.startswith('team2_')]
active_teams = [(c, X[c].values[0]) for c in team_cols if X[c].values[0] == 1]
print("\nActive team columns (should be exactly 2):")
print(active_teams)

# check venue encoding
venue_cols = [c for c in X.columns if c.startswith('venue_')]
active_venue = [(c, X[c].values[0]) for c in venue_cols if X[c].values[0] == 1]
print("\nActive venue column (should be exactly 1):")
print(active_venue)

# final prediction
prob = model.predict_proba(X)[0][1]
print(f"\nteam1 (Mumbai Indians) win prob: {prob:.1%}")
print(f"team2 (RCB) win prob: {1-prob:.1%}")
```
